#1) Quyida berilgan datasetni yuklab olib, ustun nomlarini koʻring va ularni maʼno jihatdan tushunarli shaklga keltiring. `Unnamed: 0` ustunini tashlab yuboring!

```python
import pandas as pd
import numpy as np

url = "https://raw.githubusercontent.com/alishermutalov/praktikum-datasets/refs/heads/praktikum/restaurant_sales.csv"
df = pd.read_csv(url)
df.head()
```

```python
#javobni shu yerda yozing
df.drop(columns=['Unnamed: 0'], inplace=True)

df.rename(columns = {
    'Order ID'       : 'order_id',
    'Customer ID'    : 'customer_id',
    'Category'       : 'category',
    'Item'           : 'item',
    'Price'          : 'price',
    'Quantity'       : 'quantity',
    'Order Total'    : 'order_total',
    'Order Date'     : 'order_date',
    'Payment Method' : 'payment_method'
}, inplace =True)

df.head()
```

```python
#javobni shu yerda yozing
print(df.info())
print(df.describe())
```

```python
#javobni shu yerda yozing

df1 = df.dropna(subset = ['category', 'item', 'quantity', 'order_total']).copy()
df1['payment_method'] = df1['payment_method'].fillna(df1['payment_method'].mode()[0])

#df1.isnull().sum() # checking if there is any NaN value left
df1.head()
```

#10) Boʻsh toifalar (kategoriya) mavjudligini tekshiring va toʻgʻri kategoriya berish strategiyasini ishlab chiqing.

```python
df.category.isnull().sum()
```

```python
df.groupby('category')['price'].mean()
```

```python
#javobni shu yerda yozing
def assign_category(price):
  if price <= 3:
    return 'Drinks'
  elif price <=5:
    return 'Side Dishes'
  elif price <= 5.45:
    return 'Desserts'
  elif price <= 10:
    return 'Starter'
  else:
    return 'Main Dishes'

df.loc[df['category'].isnull(), 'category'] = df.loc[df['category'].isnull(), 'price'].apply(assign_category)
df.category.isnull().sum()
```
